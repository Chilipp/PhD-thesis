Basic data visualization on a map
==========
Demo script to show all basic plot types on the map.

This example requires the psy-maps plugin and the file ```'demo.nc'``` which contains one variable for the temperature, one for zonal and one for the meridional wind direction.

````python
import psyplot.project as psy
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import cartopy.feature as cf
%matplotlib inline
````

````python
plt.rcParams['savefig.dpi'] = 300
plt.rcParams['figure.subplot.left'] = 0.12
plt.rcParams['figure.subplot.right'] = 0.92
plt.rcParams['figure.subplot.bottom'] = 0.12
plt.rcParams['figure.subplot.top'] = 0.92
plt.rcParams['figure.figsize'] = (5, 4)
````

````python
psy.rcParams['plotter.maps.xgrid'] = False
psy.rcParams['plotter.maps.ygrid'] = False
````

Visualizing scalar fields
-------------

The *mapplot* method visualizes scalar data on a map.

````python
maps = psy.plot.mapplot('psy-maps-demo.nc', name='t2m', clabel='{desc}', cmap='RdBu_r', 
                        lonlatbox=['Europe', 'Europe', 0, 'Europe'], cbar='')
pdf = maps.export('psy-maps-demo.pdf', close_pdf=False)
````

````python
maps = psy.plot.mapplot(
    'icon_grid_demo.nc', name='t2m', cmap='RdBu_r', 
    lonlatbox=['Europe', 'Europe', 0, 'Europe'], 
    datagrid=dict(color='k', linewidth=0.2), cbar='')
maps.export(pdf)
````

````python
tsunami = psy.plot.mapplot(
    'ugrid_demo.nc', name='Mesh2_height', load=True,
    maskleq=0, lonlatbox='Japan', cmap='Blues', 
    cbar='', stock_img=True, lsm='50m',
    datagrid={'c': 'k', 'lw': 0.1})
tsunami.export(pdf)
````

or via the linewidth

````python
vectors = psy.plot.mapvector(
    'psy-maps-demo.nc', name=[['u', 'v']], arrowsize=60,
    color=plt.rcParams['axes.prop_cycle'].by_key()['color'][3],
    lonlatbox=[-60, 'Europe', -10, 'Europe'], linewidth=0.005,
    map_extent=['Europe', 'Europe', 0, 'Europe'],
    arrowstyle='wedge',
    cbar='', density=(1.5, 0.70), )
vectors.plotters[0].ax.add_feature(cf.OCEAN, facecolor='0.7')
vectors.export(pdf)
````

````python
vectors = psy.plot.mapvector(
    'psy-maps-demo.nc', name=[['u', 'v']], arrowsize=None,
    color='absolute', linewidth=['absolute', 0.3], 
    lonlatbox=[-60, 'Europe', 0, 'Europe'],
    map_extent=['Europe', 'Europe', 0, 'Europe'],
    cmap='winter', cbar='', stock_img=False, plot='stream')
vectors.plotters[0].ax.add_feature(cf.OCEAN, facecolor='0.7')
vectors.export(pdf)
````

The second number for the linewidth scales the linewidth of the arrows, where the default number is 1.0

````python
combined = psy.plot.mapcombined(
    'psy-maps-demo.nc', name=[['t2m', ['u', 'v']]], cbar='', arrowsize=None, 
    cmap='binary', color=plt.rcParams['axes.prop_cycle'].by_key()['color'][3],
    lonlatbox=[-60, 'Europe', 0, 'Europe'], lsm=False,
    map_extent=['Europe', 'Europe', 0, 'Europe'], vplot='stream')
combined.plotters[0].ax.coastlines(color='w')
combined.export(pdf)
````

````python
psy.close('all')
````
